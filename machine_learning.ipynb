```python
x=df.drop('crim',axis=1)
y=df['crim']
x.head()
y.head()
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3,random_state=1)
train_x.shape
train_y.shape
test_x.shape
test_y.shape
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
```

```python
lm.fit(train_x,train_y)
```

```python
predict_test_lm=lm.predict(test_x)
```

```python
print(lm.coef_)
```

```python
df_m=pd.DataFrame({'features':x.columns,'coeff':lm.coef_})
```

```python
df_m=df_m.sort_values(by=['coeff'])
```

```python
df_m
```

```python
df_m.plot(x='features',y='coeff',kind='bar',figsize=(15,10))
plt.show();
```

```python
print("R square value for regression test data is-")
np.round(lm.score(test_x,test_y)*100,2)
```

```python
print("R square value for regression train data is-")
np.round(lm.score(train_x,train_y)*100,2)
```

```python
import numpy as np
from  sklearn import metrics
print ("simple linear regression Mean square error (MSE) for TEST data is:")
np.round(metrics.mean_squared_error(test_y,predict_test_lm),2)
```

```python
from  sklearn.metrics import mean_absolute_error
print ("simple linear regression Mean absolte error (MSE) for TEST data is:")
np.round(metrics.mean_absolute_error(test_y,predict_test_lm),2)
```
